```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

TSMC = pd.read_csv('2330stk.csv', sep='\t', index_col='Date')
TSMC.index = pd.to_datetime(TSMC.index) # 把index轉成datetime型態
Close = TSMC.Close
Open = TSMC.Open

fig = plt.figure()
plt.plot(Close['2016'], '-' , label="收盤價", marker=">")
plt.plot(Open['2016'], '-' , label="開盤價", marker="*")
plt.title('台積電 2016年收盤價/收盤價曲線',loc='right')
# loc->title的位置
plt.xlabel('日期')
plt.ylabel('收盤價')
plt.grid(True, axis='y')
plt.legend()

fig.savefig('day17_01.png')
```

```python
barValue = [0,0,0,0,0,0]
for i in Close:
    if i > 200:
        barValue[5]+=1
    if i > 190:
        barValue[4]+=1
    if i > 180:
        barValue[3]+=1
    if i > 170:
        barValue[2]+=1
    if i > 160:
        barValue[1]+=1
    else:
        barValue[0]+=1
fig = plt.figure()
plt.bar(Close['2016'], '-' , label="收盤價", marker=">")
```
